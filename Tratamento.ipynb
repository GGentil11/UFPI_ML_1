```python
import pandas as pd

df = pd.read_csv('dados_filtrados.csv')
display(df)
```

```python
df.fillna({'Certificate': 'Dado faltante'}, inplace=True)
display(df)
```

```python
def fill_missing_meta_score(row):
    if pd.isnull(row['Meta_score']):
        return row['IMDB_Rating'] * 10  # Converter a escala do IMDB_Rating para a escala do Meta_score (que Ã© de 0 a 100)
    else:
        return row['Meta_score']
```

```python
df['Meta_score'] = df.apply(fill_missing_meta_score, axis=1)

print(df.isnull().sum())

df.to_csv('dados_tratados.csv', index=False)
```
